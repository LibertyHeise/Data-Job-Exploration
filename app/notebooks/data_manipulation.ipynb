```python
# Find the average salary for each position
average_data_analyst_salary = data_analyst_salary_df['median_salary'].mean()
average_data_scientist_salary = data_scientist_salary_df['median_salary'].mean()
average_full_stack_developer_salary = full_stack_developer_salary_df['median_salary'].mean()
average_python_developer_salary = python_developer_salary_df['median_salary'].mean()
```

```python
# Make a dictionary and then covert it into a pandas dataframe 
averages_dict = {'Job Role': ['Data Analyst', 'Data Scientist', 'Full Stack Developer', 'Python Developer'], 
                 'Median Salary': [average_data_analyst_salary, average_data_scientist_salary, 
                                   average_full_stack_developer_salary, average_python_developer_salary]}

averages_df = pd.DataFrame.from_dict(averages_dict)
averages_df['Median Salary'] = averages_df['Median Salary'].astype(int)
averages_df
```

```python
# Count the amount of times certain data analysis tools appear in the job descripiton
python_counts = result_df['Job description'].str.count('Python|python').sum()
sql_counts = result_df['Job description'].str.count('SQL|sql').sum()
nosql_counts = result_df['Job description'].str.count('NoSQL|nosql').sum()
excel_counts = result_df['Job description'].str.count('Excel|excel').sum()
r_counts = result_df['Job description'].str.contains('R|r').sum()
tableau_counts = result_df['Job description'].str.count('Tableau|tableau').sum()
powerbi_counts = result_df['Job description'].str.count('Power BI|PowerBI|power bi').sum()
javascript_counts = result_df['Job description'].str.count('javascript|Javascript').sum()
html_counts = result_df['Job description'].str.count('HTML|html').sum()
css_counts = result_df['Job description'].str.count('CSS|css').sum()
mongodb_counts = result_df['Job description'].str.count('MongoDB|MONGODB').sum()
```

```python
# Make a dictionary and then convert to Data Frame
counts_dict = {'Data Analysis Tools': ['Python', 'SQL', 'NoSQL', 'Excel', 'R', 'Tableau', 'PowerBI', 'JavaScript', 'HTML', 'CSS', 'MongoDB'],
               'Count': [python_counts, sql_counts, nosql_counts, excel_counts, r_counts, tableau_counts, powerbi_counts, javascript_counts, html_counts, css_counts, mongodb_counts]}
counts_df = pd.DataFrame(counts_dict)
counts_df
```

```python
# Connect to the MongoDB server
client = MongoClient()

# Choose the database you want to use
db = client.data_jobs

# Insert the clean_states_df data frame into a collection named "states"
db.states.insert_many(states_df.to_dict('records'))

# Insert the result_df data frame into a collection named "jobs"
db.jobs.insert_many(result_df.to_dict('records'))

# Insert the averages_df into a collection named "salaries"
db.salaries.insert_many(averages_df.to_dict('records'))

# Insert the counts_df into a collection named "tools"
db.tools.insert_many(counts_df.to_dict('records'))
```
